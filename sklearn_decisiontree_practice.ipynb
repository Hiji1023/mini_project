##Decision Tree 연습##

###참고 : https://scikit-learn.org/stable/modules/tree.html#classification ###


```python
from sklearn import tree
X = [[0,0],[1,1]]
Y = [0,1]
clf = tree.DecisionTreeClassifier()
clf = clf.fit(X,Y)
```

## predict()##
###-> 각 모델의 최종적인 예측값
###R: 구체적인 특정값
###C: 확률높은 class

##predict_prob()##
###-> C에서 각 class에 대한 확률 각각 출력
###(predict probability)

```python
clf.predict([[2., 2.]])
```

```python
clf.predict_proba([[2., 2.]])
```

```python
from sklearn.datasets import load_iris
from sklearn import tree
iris = load_iris()
X,y = iris.data, iris.target
clf = tree.DecisionTreeClassifier()
clf = clf.fit(X,y)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(20,10))
t = tree.plot_tree(clf.fit(X,y), filled = True)
```
